```python
for name in sorted(os.listdir(".")):
    if not name.endswith((".html", ".meta")):
        continue
    year, month, filename = name[:4], name[4:6], name[6:]
    new_dir = "/".join([year, month])
    new_name = "/".join([new_dir, filename])
    if not os.path.exists(new_dir):
        os.makedirs(new_dir)
    if name.endswith(".html"):
        os.rename(name, new_name)
    elif name.endswith(".meta"):
        with open(name) as source:
            with open(new_name, 'w') as dest:
                for line in source:
                    if line.startswith(year + month):
                        line = line[6:]
                    dest.write(line)
        os.remove(name)
```

```python
disqus_threads = open("boredomandlaziness-2013-12-01T07:19:08.119834-links.csv").readlines()
```

```python
import re
bl_pattern = r"http://www\.boredomandlaziness\.org/(....)/(..)/(.*)"
ce_pattern = r"http://www\.curiousefficiency\.org/posts/(....)(..)(.*)"
url_map = {}
for line in disqus_threads:
    old_url = line.strip()
    if old_url == "http://www.boredomandlaziness.org/":
        new_url = "http://www.curiousefficiency.org/"
    else:
        match = re.match(ce_pattern, old_url)
        if match is None:
            match = re.match(bl_pattern, old_url)
        if match is None:
            print "Didn't match!"
            continue
        year, month, filename = match.groups([1, 2, 3])
        new_url = "http://www.curiousefficiency.org/posts/%s/%s/%s" % (year, month, filename)
    url_map[old_url] = new_url
```

```python
with open("disqus_url_map_20131201.csv", "w") as f:
    for entry in url_map.iteritems():
        f.write("%s, %s\n" % entry)
```
